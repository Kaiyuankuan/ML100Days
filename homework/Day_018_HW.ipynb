```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 設定 data_path
dir_data = './data/'
```

### 之前做過的處理

```python
# 讀取資料檔
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.shape
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])            
print(app_train.shape)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
print(app_train['DAYS_BIRTH'].describe())
app_train['YEARS_BIRTH'] = abs(app_train['DAYS_BIRTH']/365)
print(app_train['YEARS_BIRTH'].describe())
```

## 練習時間
參考 Day 17 範例程式，離散化你覺得有興趣的欄位，並嘗試找出有趣的訊息

```python
app_data = app_train[['YEARS_BIRTH','AMT_INCOME_TOTAL']]
print(app_data['AMT_INCOME_TOTAL'].describe())
bins =[20,25,30,35,40,45,50,55,60,65,70]
app_data['AGE_GROUP'] = pd.cut(app_data['YEARS_BIRTH'],bins)
print(app_data['AGE_GROUP'].value_counts())
age_income = app_data.groupby('AGE_GROUP').mean()
print(age_income['AMT_INCOME_TOTAL'])
```

```python
year_group_sorted = app_data['AGE_GROUP'].unique()
print(year_group_sorted)
app_data.boxplot(column=['AMT_INCOME_TOTAL'], by =['AGE_GROUP'], showfliers = False, figsize=(12,12))
plt.suptitle('Income vs age')
plt.show()
```
